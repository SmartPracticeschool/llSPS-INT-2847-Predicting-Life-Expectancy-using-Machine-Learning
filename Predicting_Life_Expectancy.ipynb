```python
#Transforming Data

import numpy as np
X=df.drop(['Life expectancy '],axis=1)
y=df['Life expectancy ']
```

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X=scaler.fit_transform(X);
scaler.transform(X);
```

```python
X=pd.DataFrame(X,columns=['Year', 'Status', 'Adult Mortality', 'infant deaths', 'Alcohol',
       'percentage expenditure', 'Hepatitis B', 'Measles ', ' BMI ',
       'under-five deaths ', 'Polio', 'Total expenditure', 'Diphtheria ',
       ' HIV/AIDS', 'GDP', 'Population', ' thinness  1-19 years',
       ' thinness 5-9 years', 'Income composition of resources', 'Schooling'])
```

```python
X.head()
```

```python

#removing all the column whose p-value is less than 0.05 
#after feature selection
X.drop([' thinness 5-9 years',' thinness  1-19 years'],axis=1,inplace=True)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.9, random_state=42)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X_train,y_train)
pred=lm.predict(X_test)
```

```python

from sklearn.metrics import mean_squared_error
```

```python
from math import sqrt
sqrt(mean_squared_error(y_test,pred))
```

```python
predictions = lm.predict(X_test)
errors = abs(predictions - y_test)
mape = np.mean(100 * (errors / y_test))
accuracy = 100 - mape
print('AcAcuracy:', round(accuracy, 2), '%.')
```
